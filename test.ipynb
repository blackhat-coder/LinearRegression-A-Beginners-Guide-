```python
from LinearRegression import Linear_Regression
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('headbrain.csv')
print(df.head())
```

```python
df.rename(columns = {'Head Size(cm^3)' : 'Head size' , 'Brain Weight(grams)':'Brain Weight'}, inplace=True)
print(df.columns)
```

```python
df.hist(bins=10, figsize=(20,15))
```

```python
x = df['Head size']
x = x[:, np.newaxis]
X = np.insert(x, 0, np.ones(len(df)) , axis=1)
Y = df['Brain Weight']

plt.scatter(x, Y)
plt.ylabel('Brain Weight')
plt.xlabel('Head size')
```

```python
model = Linear_Regression()
model.fit(X, Y)
```

```python
xfit = model.predict(X)

plt.scatter(x, Y)
plt.plot(x, xfit, 'k')
plt.ylabel('Brain Weight')
plt.xlabel('Head size')
plt.title('Linear model')
```
